# Feature extraction - Train

Features before clickout

```python
hotels_metadata = {}

for hotel in pd_item.values:
    item_id = hotel[0]
    properties = hotel[1].split('|')
    
    hotels_metadata[item_id] = properties
    
```

```python
## Obtaining categorical values
pd_train['platform'] = pd_train['platform'].astype("category")
pd_train['platform_cat'] = pd_train['platform'].cat.codes
lines = pd_train.values.tolist()

## Creating variables
list_features = []
n_click_out = 0
change_order = 0
search_poi = 0
search_destination = 0
n_image = []
item_info = []
item_rating = []
search_item = []
search_destination_name = []

cont = 0
Y = []

session_id = lines[0][1]
for line in lines:
    if line[4] == 'clickout item':
        
        if line[1] != session_id:
            n_click_out = 0
        if line[8] == 'tablet':
            device = 1
        elif line[8] == 'desktop':
            device = 2
        else:
            device = 3
        n_step = line[3]
        country = line[12]
        if not n_image:
            n_image.append('nan')
        if not item_rating:
            item_rating.append('nan')
        if not item_info:
            item_info.append('nan')
        if not search_item:
            search_item.append('nan')
        if not search_destination_name:
            search_destination_name.append('nan')
        prices = list(map(int,line[11].split('|')[:])) ## Converting string to int to sort
        price = list(map(int,line[11].split('|')[:]))
        price.sort()
        
        perc_of_filters = []
        items = line[10].split('|')
        if line[9] is np.nan:
            for item in items:
                perc_of_filters.append(0)
        else:
            filters = line[9].split('|')
            new_filters = []
            for f in filters:
                if f in col:
                    new_filters.append(f)
            if len(new_filters)==0:
                for item in items:
                    perc_of_filters.append(0)
            else:
                for item in items:
                    c = 0
                    if int(item) in hotels_metadata:
                        item_properties = hotels_metadata[int(item)]
                        for fil in new_filters:
                            if fil in item_properties:
                                c+=1
                        perc_of_filters.append(c/len(new_filters))
                    else:
                        perc_of_filters.append(0)
        
        for i in range(len(items)):
            if items[i] == line[5]:
                Y.append(1)
            else:
                Y.append(0)
            list_features.append([price.index(prices[i])+1,n_click_out,n_image.count(items[i]),int(n_image[-1] == items[i]),n_step,change_order,item_info.count(items[i]),int(item_info[-1] == items[i]),item_rating.count(items[i]),int(item_rating[-1] == items[i]),country,i+1,device,search_poi,search_destination,int(search_item[-1] == items[i]),int(search_destination_name[-1] == line[7]),perc_of_filters[i]])
        n_click_out += 1
        cont += 1
    else:
        if line[1] == session_id:
            if line[4] == 'interaction item image':
                n_image.append(line[5])
            elif line[4] == 'change of sort order':
                change_order = 1
            elif line[4] == 'interaction item info':
                item_info.append(line[5])
            elif line[4] == 'interaction item rating':
                item_rating.append(line[5])
            elif line[4] == 'search for poi':
                search_poi = 1
            elif line[4] == 'search for destination':
                search_destination = 1
                search_destination_name.append(line[5])
            elif line[4] == 'search for item':
                search_item.append(line[5])
        else:
            n_image = []
            change_order = 0
            search_destination = 0
            search_poi = 0
            item_info = []
            item_rating = []
            search_item = []
            search_destination_name = []
            n_click_out = 0
            if line[4] == 'interaction item image':
                n_image.append(line[5])
            elif line[4] == 'change of sort order':
                change_order = 1
            elif line[4] == 'interaction item info':
                item_info.append(line[5])
            elif line[4] == 'interaction item rating':
                item_rating.append(line[5])
            elif line[4] == 'search for poi':
                search_poi = 1
            elif line[4] == 'search for destination':
                search_destination = 1
                search_destination_name.append(line[5])
            elif line[4] == 'search for item':
                search_item.append(line[5])
    session_id = line[1]
    if cont == 50000:
        break
```

#### Hotels features

```python
# columns = ['item_id','Lift','Family Friendly','Laundry Service','Desk','Washing Machine','Fridge','Shower','Bike Rental','Childcare','Spa Hotel','Hotel Bar','Airport Shuttle','Air Conditioning','Microwave','Beach','Massage','Gym','Satellite TV','Luxury Hotel','1 Star', '2 Star', '3 Star', '4 Star', '5 Star', 'Good Rating', 'Very Good Rating', 'Satisfactory Rating', 'Excellent Rating', 'WiFi (Rooms)', 'Car Park', 'Restaurant', 'Terrace (Hotel)', 'House / Apartment']
stars = ['1 Star', '2 Star', '3 Star', '4 Star', '5 Star'] 
ratings = ['Satisfactory Rating', 'Good Rating', 'Very Good Rating', 'Excellent Rating']
hotels_stars_metadata = {}
hotels_rating_metadata = {}

for hotel in pd_item.values:
    item_id = hotel[0]
    properties = hotel[1].split('|')
    
    s = [0]
    for i in range(len(stars)):
        if stars[i] in properties:
            s.append(i+1)
    r = [0]       
    for j in range(len(ratings)):
        if ratings[j] in properties:
            r.append(j+1)
    
    hotels_stars_metadata[item_id] = max(s)
    hotels_rating_metadata[item_id] = max(r)
```

```python
len(hotels_stars_metadata)
```

```python
X_hotels = []
cont = 0

for index,row in pd_clickouts.iterrows():
    cont+=1
    items = row['impressions'].split('|')
    for item in items:
        features = []
        if int(item) in hotels_stars_metadata:
            features.append(hotels_stars_metadata[int(item)])
        else:
            features.append(0)   
            
        if int(item) in hotels_rating_metadata:
            features.append(hotels_rating_metadata[int(item)])
        else:
            features.append(0)
        
        X_hotels.append(features)
    if cont == 50000:
        break
```
