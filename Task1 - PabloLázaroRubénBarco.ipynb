# Feature extraction - Train

Features before clickout

```python
# This cell is used to obtain all the different properties of the hotels, because we want to make a new feature
# with the percentage of the properties that coincide with the filters (in each of the items in the impressions) for every 
# clickout. And we have seen that there are more filters than properties so we want to use only those ones which coincide
# with the properties of the items.
pd_item_data = pd.read_csv(dir_data+'item_dataframe.csv')
col = list(pd_item_data.columns)
del pd_item_data
```

```python
hotels_metadata = {}

for hotel in pd_item.values:
    item_id = hotel[0]
    properties = hotel[1].split('|')
    
    hotels_metadata[item_id] = properties
    
```

```python
'2 Star' in hotels_metadata[1257268]
```

```python
lines = pd_train.values.tolist()

## Creating variables
list_features = []
inter_item = []
n_click_out = 0
change_order = 0
search_poi = 0
filter_active = 0
search_destination = 0
search_destination_name = []
search_for_item = []
cont = 0
Y = []

session_id = lines[0][1]
for line in lines:
    if line[4] == 'clickout item':
        n_step = line[3]
        if not inter_item:
            inter_item.append('nan')
        if not search_destination_name:
            search_destination_name.append('nan')
        if not search_for_item:
            search_for_item.append('nan')
        prices = list(map(int,line[11].split('|')[:])) ## Converting string to int to sort
        price = list(map(int,line[11].split('|')[:]))
        price.sort()
        perc_of_filters = []
        items = line[10].split('|')
        if line[9] is np.nan:
            for item in items:
                perc_of_filters.append(0)
        else:
            filters = line[9].split('|')
            new_filters = []
            for f in filters:
                if f in col:
                    new_filters.append(f)
            if len(new_filters)==0:
                for item in items:
                    perc_of_filters.append(0)
            else:
                for item in items:
                    c = 0
                    if int(item) in hotels_metadata:
                        item_properties = hotels_metadata[int(item)]
                        for fil in new_filters:
                            if fil in item_properties:
                                c+=1
                        perc_of_filters.append(c/len(new_filters))
                    else:
                        perc_of_filters.append(0)
        
        for i in range(len(items)):
            if items[i] == line[5]:
                Y.append(1)
            else:
                Y.append(0)
            if int(items[i]) in hotels_metadata.keys():
                hotel = hotels_metadata[int(items[i])]
            else:
                hotel = [0]
            list_features.append([price.index(prices[i])+1,n_step,int(inter_item[-1] == items[i]),i+1,int(search_destination_name[-1] == line[7]),inter_item.count(items[i]),int(items[i] in search_for_item), perc_of_filters[i],int('1 Star' in hotel),int('2 Star' in hotel),int('3 Star' in hotel),int('4 Star' in hotel),int('5 Star' in hotel)])
        n_click_out += 1
        cont += 1
    else:
        if line[1] == session_id:
            if line[4] == 'interaction item image':
                inter_item.append(line[5])
            elif line[4] == 'interaction item info':
                inter_item.append(line[5])
            elif line[4] == 'interaction item rating':
                inter_item.append(line[5])
            elif line[4] == 'search for destination':
                search_destination_name.append(line[5])
            elif line[4] == 'search for item':
                inter_item.append(line[5])
                search_for_item.append(line[5])
            elif line[4] == 'interaction item deals':
                inter_item.append(line[5])
        else:
            inter_item = []
            search_for_item = []
            change_order = 0
            search_destination = 0
            filter_active = 0
            search_destination_name = []
            n_click_out = 0
            if line[4] == 'interaction item image':
                inter_item.append(line[5])
            elif line[4] == 'interaction item info':
                inter_item.append(line[5])
            elif line[4] == 'interaction item rating':
                inter_item.append(line[5])
            elif line[4] == 'search for destination':
                search_destination_name.append(line[5])
            elif line[4] == 'search for item':
                inter_item.append(line[5])
                search_for_item.append(line[5])
            elif line[4] == 'interaction item deals':
                inter_item.append(line[5])
    session_id = line[1]
    if cont == 500000:
        break
    
```
