#### Feature extraction

Features before clickout

```python
## Obtaining categorical values
pd_train['platform'] = pd_train['platform'].astype("category")
pd_train['platform_cat'] = pd_train['platform'].cat.codes
lines = pd_train.values.tolist()


## Creating variables
list_features = []
line_features = []
n_image = 0
n_click_out = 0
change_order = 0
item_info = 0
item_rating = 0
cont = 0

session_id = lines[0][1]
for line in lines:
    if line[4] == 'clickout item':
        if line[1] != session_id:
            n_click_out = 0
        if line[8] == 'tablet':
            device = 1
        elif line[8] == 'desktop':
            device = 2
        else:
            device = 3
        n_step = line[3]
        country = line[12]
        for i in range(len(line[10].split('|'))):
            list_features.append([n_click_out,n_image,n_step,device,change_order,item_info,item_rating,country])
        n_click_out += 1
        cont += 1
    else:
        if line[1] == session_id:
            if line[4] == 'interaction item image':
                n_image += 1
            elif line[4] == 'change of sort order':
                change_order += 1
            elif line[4] == 'interaction item info':
                item_info += 1
            elif line[4] == 'interaction item rating':
                item_rating += 1
        else:
            n_image = 0
            change_order = 0
            item_info = 0
            item_rating = 0
            n_click_out = 0
            if line[4] == 'interaction item image':
                n_image += 1
            elif line[4] == 'change of sort order':
                change_order += 1
            elif line[4] == 'interaction item info':
                item_info += 1
            elif line[4] == 'interaction item rating':
                item_rating += 1
    session_id = line[1]
    if cont == 50000:
        break
```

```python
len(list_features)
```

```python
X2 = np.array(list_features)
```

Features during clickout

```python
columns = ['item_id', 'Satellite TV','Luxury Hotel','1 Star', '2 Star', '3 Star', '4 Star', '5 Star', 'Good Rating', 'Very Good Rating', 'Satisfactory Rating', 'Excellent Rating', 'WiFi (Rooms)', 'Car Park', 'Restaurant', 'Terrace (Hotel)', 'House / Apartment']
new_item_metadata = {}

for hotel in pd_item.values:
    
    item_id = hotel[0]
    properties = hotel[1].split('|')
    
    info = []    
    for c in columns[1:]:
        if c in properties: 
            info.append(1)
        else:
            info.append(0)
    
    new_item_metadata[item_id] = info
```

```python
X = []
Y = []
len_col = len(columns)-1
cont = 0

for index, row in pd_clickouts.iterrows():
    cont += 1
    items = row['impressions']
    prices = row['prices']
    items = items.split('|')
    prices = prices.split('|')
    max_price = 0
    min_price = 100000000000
    clicked_out = row['reference']
    for i in range(len(items)):
        if int(prices[i]) > max_price:
            max_price = int(prices[i])
        if int(prices[i]) < min_price:
            min_price = int(prices[i])
    for i in range(len(items)):
        features_row = []
        features_row.append(i+1)
        if int(prices[i]) == max_price:
            features_row.append(1)
        else:
            features_row.append(0)
        if int(prices[i]) == min_price:
            features_row.append(1)
        else:
            features_row.append(0)
        features_row = features_row + new_item_metadata.get(int(items[i]), [0]*len_col)
        if items[i] == clicked_out:
            Y.append(1)
        else:
            Y.append(0)
        X.append(features_row)
    if cont == 50000:
        break
```
